```python
import cv2
import pandas as pd
from ultralytics import YOLO
from tracker import*

model=YOLO('yolov8s.pt')
cv2.pointPolygonTest
```

```python
tracker=Tracker()
count=0
```

```python
cap = cv2.VideoCapture('C:\\Users\\Manya\\Downloads\\Task2.mp4')
```

```python
import numpy as np
class_L = ['car', 'bus', 'truck']  


# Define polygon points 
polygon_points = np.array([[10, 30], [20, 20], [40, 20], [50, 30], [40, 40], [20, 40]], dtype=np.int32)

count = {'car': 0, 'bus': 0, 'truck': 0}

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Predict vehicle detections
    results = model.predict(frame)
    a = results[0].boxes.data.detach().cpu().numpy()
    px = pd.DataFrame(a)

    detected_vehicles = []

    for index, row in px.iterrows():
        x1, y1, x2, y2 = map(int, row[:4])
        d = int(row[5])
        c = class_L[d] if d < len(class_L) else None
        
        # Check if the center is within the polygon
        center_x = (x1 + x2) // 2
        center_y = (y1 + y2) // 2
        if cv2.pointPolygonTest(polygon_points, (center_x, center_y), False) >= 0:
            detected_vehicles.append([x1, y1, x2, y2, c])

    bbox_id = tracker.update(detected_vehicles)

    for bbox in bbox_id:
        x3, y3, x4, y4, id, c = bbox
        # Draw bounding box and label
        cv2.rectangle(frame, (x3, y3), (x4, y4), (0, 255, 0), 2)
        cv2.putText(frame, c, (x3, y3 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

        # Update counts
        if c in count:  # Ensure the vehicle class is in the count dictionary
            count[c] += 1  # Count vehicles 

    # Draw the polygon on the frame
    cv2.polylines(frame, [polygon_points], isClosed=True, color=(255, 0, 0), thickness=2)

    # Display counts
    vehicle_types = list(count.keys())
    for vehicle_type in vehicle_types:
        cv2.putText(frame, f"{vehicle_type}: {count[vehicle_type]}",
                    (10, 30 + 20 * vehicle_types.index(vehicle_type)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    # Display the frame
    cv2.imshow("Output", frame)
    if cv2.waitKey(1) & 0xFF == 27:  # Press 'Esc' to exit
        break

cap.release()
cv2.destroyAllWindows()

```
